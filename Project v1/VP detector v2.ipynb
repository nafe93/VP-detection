```python
import os
import cv2
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import cluster
from scipy.ndimage import rotate
from skimage.transform  import radon
```

```python
class VP:
    
    def __init__(self, path, cut_image=True, clossing=False):
        # answer
        self.vp = ''
        # constant
        self.clossing  = clossing
        self.cut_image = cut_image
        self.counter = 0
        self.thr1 = -1
        self.thr2 = -1
        #array
        self.points = []
        self.kernel = []
        # img
        self.img  = cv2.imread(path)
        self.img  = cv2.cvtColor(self.img, cv2.COLOR_BGR2RGB)
        self.gray = cv2.cvtColor(self.img.copy(), cv2.COLOR_RGB2GRAY)
        self.blur = ''
        self.canny= ''
        self.closing = ''
        
    def cv_view(self, img, x=7, y=7, _cmap='autumn_r'):
        if img is not None:
            plt.figure(figsize=(x, y))
            plt.imshow(img, cmap=_cmap)
            plt.show()
        else:
            print("There is no image")

    def _cv_blur_(self, kernel_x=3, kernel_y=3):
        blur = cv2.blur(self.gray,(kernel_x,kernel_y))
        self.blur = blur

    def _cv_kernel(self, n, m):
        kernel  = np.ones((n, m),np.uint8)
        return kernel

    def _cv_closing(self, img, kernel, iter_erosion= 1, iter_dilatation = 1):
        img = cv2.dilate(img, kernel, iter_dilatation)
        img = cv2.erode(img, kernel, iter_erosion)
        return img
    
    def _cv_(self, img, kernel, iter_erosion= 1, iter_dilatation = 1):
        img = cv2.dilate(img, kernel, iter_dilatation)
        img = cv2.erode(img, kernel, iter_erosion)
        return img

    def _cv_canny(self, thr1=-1, thr2=-1):
        self._cv_blur_()
        
        limit = 0.15
        
        mu, std = cv2.meanStdDev(self.blur, 0, 0)
        
        if (thr1 == -1) and (thr2 == -1):
            threshold1 = mu
            threshold2 = std
        else:
            threshold1 = mu  - thr1
            threshold2 = std - thr2

        print(threshold1, threshold2)
        self.canny = cv2.Canny(self.blur, threshold1, threshold2)    
        
        if self.cut_image==True:
                        
            h, w = self.blur.shape
            
            limit = 0.075
            
            region_of_VP = [(0, h), (w/2, 0), (w, h)]
            region_of_VP = np.array([region_of_VP], np.int32)
            
            region_of_VP_size = len(region_of_VP.reshape(-1))
            
            mask = np.zeros_like(self.blur)
            match_mask_color = 255

            cv2.fillPoly(mask, region_of_VP, match_mask_color)
            masked_image = cv2.bitwise_and(self.canny, mask)

            self.canny = masked_image

        if self.clossing == True:
            if len(self.kernel) == 0:
                self.kernel = self._cv_kernel(1, 3)
#             self.canny = self._cv_closing(masked_image, self.kernel)
            self.canny = cv2.dilate(self.canny, self._cv_kernel(3, 1))
        
        
        count_white_element = np.unique(self.canny, return_counts=True)[1][1]

        canny_img = len(self.canny.reshape(-1)) 
        percent   = count_white_element / canny_img
        if percent < limit:
            if self.counter < 500:
                self.thr1 += 2
                self.thr2 += 1
                self.counter += 1
                self._cv_canny(self.thr1, self.thr2)
            else:
                print(percent)
                    
#         def radon_transform(self):
            
```

```python
def cluster_intersects(points):
    db = cluster.KMeans(eps=10, min_samples=1).fit(np.array(points))
    labels = db.labels_
    biggest_cluster = np.argmax(np.bincount(labels))
    vp = tuple(np.mean(np.array(points)[labels==biggest_cluster], axis=0).astype('int'))
    return vp
```

```python
def draw_vp_of_clustering(img, points):
    new_img = img.copy()
    vp = cluster_intersects(points)
    new_img = cv2.drawMarker(new_img, (vp), (0, 255, 255), cv2.MARKER_CROSS, 5, 5, 1)
    print(vp)
    return new_img
```
