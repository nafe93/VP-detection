```python
import os
import cv2
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import cluster
```

```python
class VP:
    
    def __init__(self, img_path, cut_image=True, houghlines_divisor=15):
        #vp 
        self.vp = ''
        #bool
        self.stop = False
        self.cut_image = cut_image
        #parametrs
        self.hough_threshold    = 0
        self.min_crossing_angle = 1
        self.houghlines_divisor = houghlines_divisor
        self.points  = []
        self.lines   = []
        self.counter = 0
        self.thr1 = -1
        self.thr2 = -1
        #image
        self.img   = cv2.imread(img_path)
        self.img   = cv2.cvtColor(self.img, cv2.COLOR_BGR2RGB)
        self.gray  = cv2.cvtColor(self.img, cv2.COLOR_RGB2GRAY)
        self.blur  = ''
        self.canny = ''
        self.img_l = ''
    
    def error(self):
        if self.counter < 250:
#             print(f"{self.counter}/250")
            if self.counter == 0:
                self.thr1 += 2
                self.thr2 += 2
            elif self.counter > 0:
                self.thr1 += 1
                self.thr2 += 1

            self.counter += 1
            self.draw_hough_lines(self.img)
        else:
            print("We can't find line in image")
            self.stop = True
            
    
    def cv_view(self, img, x=7, y=7, _cmap='autumn_r'):
        if img is not None:
            plt.figure(figsize=(x, y))
            plt.imshow(img, cmap=_cmap)
            plt.show()
        else:
            print("There is no image")       

    def _cv_blur_(self, kernel_x=3, kernel_y=3):
        blur = cv2.blur(self.gray,(kernel_x,kernel_y))
        self.blur = blur
    
    def cv_kernel(self, n, m):
        kernel  = np.ones((n, m),np.uint8)
        return kernel
    
    def cv_closing(self, img, kernel, iter_erosion= 1, iter_dilatation = 1):
        img = cv2.dilate(img, kernel, iter_dilatation)
        img = cv2.erode(img, kernel, iter_erosion)
        return img
    
    def _cv_canny(self, thr1=-1, thr2=-1):
        self._cv_blur_()
        mu, std = cv2.meanStdDev(self.blur, 0, 0)
        
        if (thr1 == -1) and (thr2 == -1):
            threshold1 = mu
            threshold2 = std
        else:
            threshold1 = mu  - thr1
            threshold2 = std - thr2
            
        self.canny = cv2.Canny(self.blur, threshold1, threshold2)
        
        if self.cut_image==True:
            
            h, w = self.blur.shape
            region_of_VP = [(0, h), (w/2, h/2), (w, h)]
            region_of_VP = np.array([region_of_VP], np.int32)
            
            mask = np.zeros_like(self.blur)
            match_mask_color = 255
    
            cv2.fillPoly(mask, region_of_VP, match_mask_color)
            masked_image = cv2.bitwise_and(self.canny, mask)
            
            kernel = self.cv_kernel(1, 1)
            
            self.canny = self.cv_closing(masked_image, kernel)
            
            
        
    def calculate_hough_threshold_(self, houghlines_divisor):
        self._cv_canny(thr1=self.thr1, thr2=self.thr2)
        
        image_rows = self.gray.shape[0] ** 2
        image_cols = self.gray.shape[1] ** 2
    
        hough_threshold = (np.sqrt(image_rows + image_cols) / houghlines_divisor)
    
        return hough_threshold 
    
    def _hough_threshold_(self):
        self.hough_threshold =  int(self.calculate_hough_threshold_(self.houghlines_divisor))
    
    
    def _cv_lines_(self):
        self._hough_threshold_()
        lines  = cv2.HoughLines(self.canny, 1, np.pi / 180, self.hough_threshold)
        self.lines = lines
    
    
    def draw_hough_lines(self, img):
        self._cv_lines_()

        if (self.lines is not None) and (self.lines.size > 2):
            copy_img = img.copy()
            img2 = img.copy()

            for i in range(len(self.lines)):
                # rho Это дистанция до кординаты (center,center)
                # theta угол повората изображения
                rho    = self.lines[i][0][0]
                theta1 = self.lines[i][0][1]

                theta_thresh  = 0.2

                costheta1 = 0
                costheta2 = 0

                sintheta1 = 0
                sintheta2 = 0 

                rho1 = 0
                rho2 = 0

                if ((theta1 >(0 + theta_thresh)) and (theta1 < (np.pi / 2 - theta_thresh)) or (theta1 >((np.pi / 2) + theta_thresh)) and (theta1 < (np.pi - theta_thresh))):

                    # a нормирование 
                    # b нормирование
                    a = np.cos(theta1)
                    b = np.sin(theta1)
                    # кадинаты
                    x0 = a * rho
                    y0 = b * rho

                    # начало вектора
                    x1 = int(x0 + 1000 * (-b))
                    y1 = int(y0 + 1000 * (a))
                    # конец вектора
                    x2 = int(x0 - 1000 * (-b))
                    y2 = int(y0 - 1000 * (a))
                    # Рисование 
                    cv2.line(copy_img, (x1, y1), (x2, y2), (0, 255, 255), 2)

                    costheta1 = np.cos(theta1)
                    sintheta1 = np.sin(theta1)
                    rho1 = rho

                    for j in range(i + 1, len(self.lines)):
                        rho    = self.lines[j][0][0]
                        theta2 = self.lines[j][0][1]

                        costheta2 = np.cos(theta2)
                        sintheta2 = np.sin(theta2)

                        rho2 = rho

                        if ((np.abs(theta1) - np.abs(theta2)) < self.min_crossing_angle):
                            break

                        determinative = (costheta1 * sintheta2) - (sintheta1 * costheta2)

                        intersection_x = ( ( sintheta2 * rho1) - (sintheta1 * rho2) ) / determinative
                        intersection_y = ( (-costheta2 * rho1) + (costheta1 * rho2) ) / determinative

                        if (intersection_x > 0) and (intersection_y > 0):
                            if not np.isinf(intersection_x):
                                if not np.isinf(intersection_y):

                                    intersection_x = int(intersection_x)
                                    intersection_y = int(intersection_y)
                                    self.points.append([intersection_x, intersection_y])
                                    img2 = cv2.drawMarker(img2, (intersection_x, intersection_y), (0, 255, 255), cv2.MARKER_CROSS, 5, 5, 1)
            self.img_l = img2
        else:
            if self.stop == False:
                self.error()
                
    
    def _cluster_intersects(self, eps=5):
        if len(self.points) > 250:
            db = cluster.DBSCAN(eps=eps, min_samples=1).fit(np.array(self.points))
            labels = db.labels_
            biggest_cluster = np.argmax(np.bincount(labels))
            vp = tuple(np.mean(np.array(self.points)[labels==biggest_cluster], axis=0).astype('int'))
            if vp:
                self.vp = vp
            else:
                if self.stop == False:
                    self.error()
                    self._cluster_intersects()
                    self.points = []
        else:
            if self.stop == False:
                self.error()
                self._cluster_intersects()
                self.points = []
    
    def draw_vp_of_clustering(self, eps=5):
        new_img = self.img.copy()
        self._cluster_intersects(eps=eps)
        if self.vp:
            new_img = cv2.drawMarker(new_img, self.vp, (0, 255, 255), cv2.MARKER_CROSS, 5, 5, 1)
            return new_img
        else:
            print("We can't find VP in image")
```

```python
images = [i for i in os.listdir('../test_1/test_1/') if 'jpg' in i]
data = []
```

```python
for index, image in enumerate(images):
    vp = VP('../test_1/test_1/'+ image)
    vp.draw_hough_lines(vp.img)
    vp_img = vp.draw_vp_of_clustering(eps=5)
    vp.cv_view(vp.canny, _cmap='gray')
    vp.cv_view(vp_img)
    data.append([image, vp.vp])
```
