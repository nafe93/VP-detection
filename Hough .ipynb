```python
import numpy as np 
import cv2

%matplotlib inline
import matplotlib.pyplot as plt

import pandas as pd
import os
```

```python
img   = cv2.imread('sudoku.png')
gray  = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
canny = cv2.Canny(gray, 50 , 150, apertureSize=3)
lines = cv2.HoughLines(canny, 1, np.pi / 180, 200) # Последние значение порог
# т.е. только лини которые имееют достаточную силу
```

```python
for line in lines:
    # rho Это дистанция до кординаты (center,center)
    # theta угол повората изображения
    rho, theta = line[0]
    
    # a нормирование 
    # b нормирование
    a = np.cos(theta)
    b = np.sin(theta)
    # кадинаты
    x0 = a * rho
    y0 = b * rho
    # начало вектора
    x1 = int(x0 + 1000 * (-b))
    y1 = int(y0 + 1000 * (a))
    # конец вектора
    x2 = int(x0 - 1000 * (-b))
    y2 = int(y0 - 1000 * (a))
    # Рисование 
    cv2.line(img, (x1, y1), (x2, y2), (0, 255, 255), 2)
```

```python
plt.imshow(img)
```

# Probablistic Hough Line Transform

```python
img   = cv2.imread('./sudoku.png')
gray  = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
canny = cv2.Canny(gray, 50 , 150, apertureSize=3)
lines = cv2.HoughLinesP(canny, 1, np.pi / 180, 100, minLineLength=250, maxLineGap=20)
# minLineLength: минимальная длина лини
# maxLineGap: максимальный разрыв
```

```python
for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(img, (x1, y1), (x2, y2), (0, 255, 255), 2)
```

```python
plt.imshow(img)
plt.show()
plt.imshow(canny)
```
